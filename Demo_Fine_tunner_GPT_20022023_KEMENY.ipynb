```python
import json
import openai
import requests
from google.colab import files
import os
import re
import time

#Agrega tu API KEY
openai.api_key = "API KEY"

# import openai_secret_manager
```

```python
# Carga el documento en la carpeta local de Colab, 
with open('/content/linkedinpost.txt', 'r') as f:
    text = f.read()

# Separa el texto en frases
sentences = text.split('. ')

# Crea una lista de archivos JSONL
json_objects = []
for sentence in sentences:
    json_objects.append({'text': sentence.strip()})

# Escribe el JSONL
with open('/content/mytext.jsonl', 'w') as f:
    for json_obj in json_objects:
        json.dump({'prompt': '', 'completion': json_obj['text']}, f)
        f.write('\n')

# Carga el JSONL en la memoria
with open('/content/mytext.jsonl', 'r') as f:
    file_content = f.read()

# Crea un objeto para el Fine-Tune
response = openai.File.create(purpose="fine-tune", file=file_content)

# Crea el Fine-Tune Job, elige el modelo a usar. ADA es el más liviano en costos.
model = "ada"
training_file = response.id
fine_tune = openai.FineTune.create(
    model=model,
    training_file=training_file,
    n_epochs=4,
    batch_size=32,
    prompt_loss_weight=0.01
)
print(f"Fine-tuning job created. Job ID: {fine_tune.id}")
```

```python

# Check the status of the fine-tuning job
job = openai.FineTune.retrieve(fine_tune.id)
print(f"Fine-tuning job status: {job.status}")

# Wait for the fine-tuning job to complete
while job.status != "succeeded":
    job = openai.FineTune.retrieve(fine_tune.id)
    time.sleep(10)

print(f"Fine-tuning job status: {job.status}")

# Print the job object to check if it is valid
print(job)

# Print the URL to the fine-tuned model
model_id = job.model
print(f"Fine-tuned model created. Model ID: {model_id}")
```

```python
# Consulta tu modelo
prompt = input("Enter your prompt: ")
response = openai.Completion.create(
    engine=model_id,
    prompt=prompt,
    max_tokens=50,
)
print(response.choices[0].text)

response = openai.Completion.create(
    prompt=prompt,
    max_tokens=500,
    n=1,
    stop=None,
    temperature=0.7,
    model=model_id,
)

# Imprime respuesta
print(response.choices[0].text)
```
